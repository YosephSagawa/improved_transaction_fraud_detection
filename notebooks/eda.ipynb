```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load data

```python
fraud_data = pd.read_csv('../data/raw/Fraud_Data.csv')
creditcard_data = pd.read_csv('../data/raw/creditcard.csv')
```

## Univariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(fraud_data['purchase_value'], bins=30)
plt.title('Distribution of Purchase Value')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='source', hue='class', data=fraud_data)
plt.title('Fraud by Source')
plt.show()
```

## Class Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=fraud_data)
plt.title('Class Distribution in Fraud_Data')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=creditcard_data)
plt.title('Class Distribution in creditcard')
plt.show()
```
